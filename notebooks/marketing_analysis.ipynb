<h3>⭐ Customer Acquisition Report<h3>

```python
#Revenue and Orders From Conversions
query_model_mart_campaign_orders = """
CREATE OR REPLACE TABLE model_mart_campaign_orders AS
SELECT
    ci.campaign_id,
    count(DISTINCT ot.order_id) as orders,
    count(DISTINCT ot.customer_id) as customers_acquired,
    SUM(ot.total_amount) as revenue
FROM order_transactions ot
JOIN campaign_interactions ci ON ci.order_id = ot.order_id
WHERE campaign_id IS NOT NULL
GROUP BY campaign_id;
"""
con.execute(query_model_mart_campaign_orders)
```

```python
#Number of first-time customers attributed to a specific campaign.
query_model_stg_first_orders  = """
CREATE OR REPLACE TABLE model_stg_first_orders  AS
SELECT
    customer_id,
    MIN(order_datetime) as first_order_time
FROM order_transactions
GROUP BY customer_id;

CREATE OR REPLACE TABLE model_mart_customer_acquisition AS
SELECT 
    o.customer_id,
    ci.campaign_id,
    o.order_id,
    o.order_datetime,
    o.total_amount
FROM order_transactions o
JOIN campaign_interactions ci ON ci.order_id = o.order_id
JOIN model_stg_first_orders ms  ON o.customer_id = ms.customer_id AND o.order_datetime = ms.first_order_time
WHERE ms.customer_id IS NOT NULL
"""
con.execute(query_model_stg_first_orders )
```

```python
#Spending behavior of those customers
query_model_mart_customer_lifetime_orders   = """
CREATE OR REPLACE TABLE model_mart_customer_lifetime_orders   AS
SELECT
    ot.customer_id,
    ci.campaign_id,
    SUM(ot.total_amount) AS total_spent,
    AVG(ot.total_amount) AS average_purchase_value,
    COUNT(ot.order_id) AS total_orders
FROM order_transactions ot
JOIN campaign_interactions ci ON ot.customer_id = ci.customer_id
GROUP BY ot.customer_id,ci.campaign_id
"""
con.execute(query_model_mart_customer_lifetime_orders)
```

```python
#Average time between first interaction and first purchase
query_model_mart_customer_interaction_to_order = """
CREATE OR REPLACE TABLE model_mart_customer_interaction_to_order AS
SELECT
    ot.customer_id,
    ci.campaign_id,
    MIN(ot.order_datetime) AS first_order_time,
    MIN(ci.interaction_datetime) AS first_interaction_time,
    DATE_DIFF('hour',MIN(ot.order_datetime) ,MIN(ci.interaction_datetime)) AS time_to_convert_hr
FROM order_transactions ot
JOIN campaign_interactions ci ON ot.customer_id = ci.customer_id
GROUP BY ot.customer_id,ci.campaign_id
"""
con.execute(query_model_mart_customer_interaction_to_order)
```

```python
# How long customers remain active after their first purchase. 
query_model_mart_customer_lifetime    = """
CREATE OR REPLACE TABLE model_mart_customer_lifetime    AS
SELECT
    ot.customer_id,
    ci.campaign_id,
    MIN(ot.order_datetime) AS first_order_time,
    MAX(ot.order_datetime) AS last_order_time,
    DATE_DIFF('day', MIN(ot.order_datetime), MAX(ot.order_datetime)) AS active_days
FROM order_transactions ot
JOIN campaign_interactions ci ON ot.customer_id = ci.customer_id
GROUP BY ot.customer_id,ci.campaign_id
"""
con.execute(query_model_mart_customer_lifetime )
```

```python
query_report_customer_acquisition  = """
CREATE OR REPLACE TABLE report_customer_acquisition AS
SELECT
    acq.customer_id,
    acq.campaign_id,
    
    -- Spending behavior
    lifetime.total_orders,
    lifetime.total_spent,
    lifetime.average_purchase_value,

    -- Active duration
    act.first_order_time,
    act.last_order_time,
    act.active_days,

    -- Time to convert
    inter.first_interaction_time,
    inter.time_to_convert_hr

FROM model_mart_customer_acquisition acq
LEFT JOIN model_mart_customer_lifetime_orders lifetime
    ON acq.customer_id = lifetime.customer_id AND acq.campaign_id = lifetime.campaign_id
LEFT JOIN model_mart_customer_lifetime act
    ON acq.customer_id = act.customer_id AND acq.campaign_id = act.campaign_id
LEFT JOIN model_mart_customer_interaction_to_order inter
    ON acq.customer_id = inter.customer_id AND acq.campaign_id = inter.campaign_id;
"""
con.execute(query_report_customer_acquisition)
```

```python
query_report_retargeting_campaigns = """
WITH retargeting_campaigns AS (
    SELECT 
        campaign_id, campaign_name
    FROM campaign_master
    WHERE LOWER(campaign_type) LIKE '%retarget%'
),

model_customer_retargeting_campaign AS (
    SELECT DISTINCT ci.customer_id, ci.campaign_id, ci.interaction_datetime
    FROM campaign_interactions ci
    JOIN campaign_master cm ON ci.campaign_id = cm.campaign_id
    WHERE LOWER(campaign_type) LIKE '%retarget%'
),

model_customer_retargeting_previously_active AS (
    SELECT 
        COUNT(DISTINCT ci.customer_id) as number_customer_previously_active
    FROM campaign_interactions ci
    JOIN campaign_master cm ON ci.campaign_id = cm.campaign_id
    JOIN order_transactions ot ON ci.customer_id = ot.customer_id
    WHERE LOWER(cm.campaign_type) LIKE '%retarget%'
        AND ot.order_datetime < ci.interaction_datetime
),

model_returning_customers AS (
    SELECT 
        ci.customer_id, ci.campaign_id, MIN(ot.order_datetime) AS first_return_time
    FROM campaign_interactions ci
    JOIN campaign_master cm ON ci.campaign_id = cm.campaign_id
    JOIN order_transactions ot ON ci.customer_id = ot.customer_id
    WHERE LOWER(cm.campaign_type) LIKE '%retarget%'
        AND ot.order_datetime > ci.interaction_datetime
    GROUP BY ci.customer_id, ci.campaign_id
),

model_total_spend_retargeted AS (
    SELECT 
        ci.customer_id, ci.campaign_id, SUM(ot.total_amount) AS total_spend_after
    FROM campaign_interactions ci
    JOIN campaign_master cm ON ci.campaign_id = cm.campaign_id
    JOIN order_transactions ot ON ci.customer_id = ot.customer_id
    WHERE LOWER(cm.campaign_type) LIKE '%retarget%'
        AND ot.order_datetime > ci.interaction_datetime
    GROUP BY ci.customer_id, ci.campaign_id
),

model_time_gap_order AS (
    SELECT 
        ci.customer_id, ci.campaign_id,
        MIN(CASE WHEN ot.order_datetime < ci.interaction_datetime THEN ot.order_datetime END) AS last_pre_order,
        MIN(CASE WHEN ot.order_datetime > ci.interaction_datetime THEN ot.order_datetime END) AS first_post_order,
        DATE_DIFF('day',
            MIN(CASE WHEN ot.order_datetime < ci.interaction_datetime THEN ot.order_datetime END),
            MIN(CASE WHEN ot.order_datetime > ci.interaction_datetime THEN ot.order_datetime END)
        ) AS gap_day
    FROM campaign_interactions ci
    JOIN campaign_master cm ON ci.campaign_id = cm.campaign_id
    JOIN order_transactions ot ON ci.customer_id = ot.customer_id
    WHERE LOWER(cm.campaign_type) LIKE '%retarget%'
    GROUP BY ci.customer_id, ci.campaign_id
),

model_retention_retargeting AS (
    SELECT 
        ci.customer_id, ci.campaign_id,
        COUNT(ot.order_id) AS num_orders_retention,
        MAX(ot.order_datetime) AS last_active
    FROM campaign_interactions ci
    JOIN campaign_master cm ON ci.campaign_id = cm.campaign_id
    JOIN order_transactions ot ON ci.customer_id = ot.customer_id
    WHERE LOWER(cm.campaign_type) LIKE '%retarget%'
        AND ot.order_datetime > ci.interaction_datetime
    GROUP BY ci.customer_id, ci.campaign_id
)

SELECT 
    i.customer_id,
    i.campaign_id,
    rc.campaign_name,
    i.interaction_datetime,
    rcust.first_return_time,
    spend.total_spend_after,
    gap.last_pre_order,
    gap.first_post_order,
    gap.gap_day,
    ret.num_orders_retention,
    ret.last_active
FROM model_customer_retargeting_campaign i
JOIN retargeting_campaigns rc ON i.campaign_id = rc.campaign_id
LEFT JOIN model_returning_customers rcust ON i.customer_id = rcust.customer_id AND i.campaign_id = rcust.campaign_id
LEFT JOIN model_total_spend_retargeted spend ON i.customer_id = spend.customer_id AND i.campaign_id = spend.campaign_id
LEFT JOIN model_time_gap_order gap ON i.customer_id = gap.customer_id AND i.campaign_id = gap.campaign_id
LEFT JOIN model_retention_retargeting ret ON i.customer_id = ret.customer_id AND i.campaign_id = ret.campaign_id
ORDER BY i.interaction_datetime;
"""
df_report_retargeting_campaigns = pd.read_sql_query(query_report_retargeting_campaigns, con)
df_report_retargeting_campaigns
```

<h3> ⭐ Driver Performance Report<h3>

```python
query_report_driver_performance = """
WITH complete_order AS (
    SELECT
        ot.order_id,
        ot.driver_id,
        ot.order_datetime,
        ot.delivery_datetime,
        DATEDIFF('minute', ot.order_datetime, ot.delivery_datetime) AS delivery_minutes,
        CASE 
            WHEN ot.delivery_datetime > ot.pickup_datetime THEN 1 
            ELSE 0 
        END AS is_late
    FROM order_transactions ot
    WHERE ot.order_status = 'completed'
),

tasks_summary AS (
    SELECT
        driver_id,
        COUNT(*) AS tasks_assigned,
        SUM(CASE WHEN order_status = 'completed' THEN 1 ELSE 0 END) AS tasks_completed
    FROM order_transactions
    GROUP BY driver_id
),

delivery_stats AS (
    SELECT
        driver_id,
        COUNT(*) AS completed_count,
        ROUND(AVG(delivery_minutes), 2) AS avg_delivery_time_mins,
        SUM(is_late) AS late_deliveries
    FROM complete_order
    GROUP BY driver_id
),

feedback AS (
    SELECT
        st.driver_id,
        COUNT(*) AS complaints
    FROM support_tickets st
    WHERE st.issue_type = 'driver'
    GROUP BY st.driver_id
),

final AS (
    SELECT
        d.driver_id,
        d.vehicle_type,
        d.region,
        ts.tasks_assigned,
        ts.tasks_completed,
        ROUND(100.0 * ts.tasks_completed / NULLIF(ts.tasks_assigned, 0), 2) AS acceptance_rate_pct,
        ds.avg_delivery_time_mins,
        ds.late_deliveries,
        COALESCE(f.complaints, 0) AS customer_complaints
    FROM drivers_master d
    LEFT JOIN tasks_summary ts ON d.driver_id = ts.driver_id
    LEFT JOIN delivery_stats ds ON d.driver_id = ds.driver_id
    LEFT JOIN feedback f ON d.driver_id = f.driver_id
)

SELECT *
FROM final
ORDER BY acceptance_rate_pct DESC NULLS LAST;
"""

df_report_driver_performance = pd.read_sql_query(query_report_driver_performance, con)
df_report_driver_performance
```

<h3> ⭐ Delivery Zone Heatmap Report<h3>

```python
query_report_delivery_zone_heatmap = """
WITH zone_orders AS (
    SELECT 
        delivery_zone AS zone,
        COUNT(*) AS total_request_order,
        SUM(CASE WHEN order_status = 'completed' THEN 1 ELSE 0 END) AS completed_orders,
        SUM(CASE WHEN order_status IN ('failed', 'cancelled') THEN 1 ELSE 0 END) AS failed_orders,
        ROUND(AVG(
            CASE 
                WHEN order_status = 'completed' THEN DATEDIFF('minute', order_datetime, delivery_datetime) 
                ELSE NULL 
            END
        ), 2) AS avg_delivery_time_mins,
        ROUND(AVG(
            CASE 
                WHEN order_status = 'completed' THEN DATEDIFF('minute', order_datetime, pickup_datetime)
                ELSE NULL
            END
        ), 2) AS avg_pickup_time_mins
    FROM order_transactions
    GROUP BY delivery_zone
),

zone_drivers AS (
    SELECT 
        ot.delivery_zone AS zone,
        COUNT(*) AS drivers_available
    FROM drivers_master dm
    JOIN order_transactions ot ON ot.driver_id = dm.driver_id
    GROUP BY ot.delivery_zone
),

Final_report AS (
    SELECT
        z.zone,
        z.total_request_order,
        z.completed_orders,
        ROUND(100.0 * z.completed_orders / NULLIF(z.total_request_order, 0), 2) AS completion_rate_pct,
        z.failed_orders,
        z.avg_delivery_time_mins,
        z.avg_pickup_time_mins,
        zd.drivers_available,
        ROUND(CAST(zd.drivers_available AS FLOAT) / NULLIF(z.total_request_order, 0), 2) AS driver_to_request_ratio,
        ROUND(z.avg_delivery_time_mins - z.avg_pickup_time_mins, 2) AS delivery_expectation_gap
    FROM zone_orders z
    LEFT JOIN zone_drivers zd ON z.zone = zd.zone
)

SELECT * FROM Final_report
ORDER BY completion_rate_pct ASC;
"""


df_report_delivery_zone_heatmap = pd.read_sql_query(query_report_delivery_zone_heatmap, con)
df_report_delivery_zone_heatmap
```

<h3> ⭐ Driver Incentive Impact Report<h3>

```python
query_driver_incentive_impact = """
WITH incentive_summary AS (
    SELECT
        oli.driver_id,
        oli.incentive_program,
        CAST(oli.applied_date AS DATE) AS applied_date,
        oli.bonus_amount,
        oli.actual_deliveries,
        oli.delivery_target,
        oli.bonus_qualified
    FROM order_log_incentive_sessions_driver_incentive_logs oli
),

incentive_orders AS (
    SELECT
        ot.driver_id,
        CAST(ot.order_datetime AS DATE) AS delivery_date,
        ROUND(AVG(DATEDIFF('minute', ot.order_datetime, ot.delivery_datetime)), 2) AS avg_delivery_time,
        COUNT(*) FILTER (WHERE ot.order_status = 'completed') AS completed_orders,
        COUNT(*) FILTER (WHERE ot.order_status IN ('failed', 'cancelled')) AS failed_or_cancelled
    FROM order_transactions ot
    GROUP BY ot.driver_id, CAST(ot.order_datetime AS DATE)
),

feedback_during_incentive AS (
    SELECT
        st.driver_id,
        CAST(st.opened_datetime AS DATE) AS feedback_date,
        COUNT(*) AS total_feedback
    FROM support_tickets st
    WHERE st.driver_id IS NOT NULL
    GROUP BY st.driver_id, CAST(st.opened_datetime AS DATE)
),

incentive_report AS (
    SELECT 
        i.driver_id,
        dm.driver_rating,
        dm.vehicle_type,
        i.incentive_program,
        i.applied_date,
        i.bonus_amount,
        i.actual_deliveries,
        i.delivery_target,
        i.bonus_qualified,
        io.completed_orders,
        io.failed_or_cancelled,
        ROUND(100.0 * io.completed_orders / NULLIF(io.completed_orders + io.failed_or_cancelled, 0), 2) AS acceptance_rate_pct,
        io.avg_delivery_time,
        COALESCE(fd.total_feedback, 0) AS feedback_count
    FROM incentive_summary i
    LEFT JOIN incentive_orders io ON i.driver_id = io.driver_id AND i.applied_date = io.delivery_date
    LEFT JOIN feedback_during_incentive fd ON i.driver_id = fd.driver_id AND i.applied_date = fd.feedback_date
    LEFT JOIN drivers_master dm ON i.driver_id = dm.driver_id
)

SELECT *
FROM incentive_report
ORDER BY bonus_amount DESC NULLS LAST;
"""

df_report_driver_incentive_impact = pd.read_sql_query(query_driver_incentive_impact, con)
df_report_driver_incentive_impact
```

<h3> ⭐ Complaint Summary Dashboard<h3>

```python
query_complaint_summary = """
WITH issue_summary AS (
    SELECT 
        CAST(opened_datetime AS DATE) AS issue_date,
        issue_type,
        issue_sub_type,
        status,
        ticket_id,
        DATEDIFF('minute', opened_datetime, resolved_datetime) AS resolution_time_min,
        compensation_amount
    FROM support_tickets
),

issue_aggregates AS (
    SELECT 
        COUNT(*) AS total_issue,
        COUNT(*) FILTER (WHERE status NOT IN ('resolved', 'closed')) AS unresolved_or_escalated,
        ROUND(AVG(resolution_time_min), 2) AS avg_resolution_time_min,
        SUM(COALESCE(compensation_amount, 0)) AS total_compensation
    FROM issue_summary
),

top_categories AS (
    SELECT
        issue_type,
        COUNT(*) AS issue_count
    FROM issue_summary
    GROUP BY issue_type
    ORDER BY issue_count DESC
    LIMIT 5
),

daily_trends AS (
    SELECT
        issue_date,
        COUNT(*) AS issue_count,
        ROUND(AVG(resolution_time_min), 2) AS avg_resolution_time_min
    FROM issue_summary
    GROUP BY issue_date
    ORDER BY issue_date
)

SELECT 
    ia.total_issue,
    ia.unresolved_or_escalated,
    ia.avg_resolution_time_min,
    ia.total_compensation,
    tc.issue_type AS top_issue_type,
    tc.issue_count AS top_issue_count,
    dt.issue_date,
    dt.issue_count AS daily_issue_count,
    dt.avg_resolution_time_min AS daily_avg_resolution
FROM issue_aggregates ia
LEFT JOIN top_categories tc ON TRUE
LEFT JOIN daily_trends dt ON TRUE;

"""


df_complaint_summary = pd.read_sql_query(query_complaint_summary, con)
df_complaint_summary
```

<h3> ⭐ Driver-Related Complaints Report<h3>

```python
query_driver_complaint_report = """
WITH driver_issues AS (
    SELECT 
        st.driver_id,
        st.issue_type,
        st.issue_sub_type,
        DATEDIFF('minute', st.opened_datetime, st.resolved_datetime) AS resolution_time_min,
        st.ticket_id,
        st.resolved_datetime,
        st.opened_datetime
    FROM support_tickets st
    WHERE st.driver_id IS NOT NULL
),

complaint_counts AS (
    SELECT
        driver_id,
        COUNT(*) AS total_complaints,
        ROUND(AVG(resolution_time_min), 2) AS avg_resolution_time_min
    FROM driver_issues
    GROUP BY driver_id
),

complaint_types AS (
    SELECT 
        driver_id,
        issue_sub_type,
        COUNT(*) AS type_count
    FROM driver_issues
    GROUP BY driver_id, issue_sub_type
),

driver_order_counts AS (
    SELECT
        driver_id,
        COUNT(*) AS total_orders
    FROM order_transactions
    WHERE order_status = 'completed'
    GROUP BY driver_id
),

complaint_ratio AS (
    SELECT 
        cc.driver_id,
        cc.total_complaints,
        cc.avg_resolution_time_min,
        doc.total_orders,
        ROUND(CAST(cc.total_complaints AS FLOAT) / NULLIF(doc.total_orders, 0), 2) AS complaint_to_order_ratio
    FROM complaint_counts cc
    LEFT JOIN driver_order_counts doc ON cc.driver_id = doc.driver_id
),

driver_ratings_change AS (
    SELECT
        driver_id,
        driver_rating AS current_rating
    FROM drivers_master
)

SELECT 
    cr.driver_id,
    cr.total_complaints,
    cr.total_orders,
    cr.complaint_to_order_ratio,
    cr.avg_resolution_time_min,
    drc.current_rating,
    ct.issue_sub_type,
    ct.type_count
FROM complaint_ratio cr
LEFT JOIN driver_ratings_change drc ON cr.driver_id = drc.driver_id
LEFT JOIN complaint_types ct ON cr.driver_id = ct.driver_id
ORDER BY complaint_to_order_ratio DESC NULLS LAST;
"""
df_driver_complaint_report = pd.read_sql_query(query_driver_complaint_report,con)
df_driver_complaint_report

```

<h3> ⭐ Restaurant Quality Complaint Report<h3>

```python
query_restaurant_complaint_report = """
WITH restaurant_complaints AS (
    SELECT
        st.restaurant_id,
        st.issue_type,
        st.issue_sub_type,
        DATEDIFF('minute', st.opened_datetime, st.resolved_datetime) AS resolution_time_min,
        st.compensation_amount,
        st.order_id,
        ot.customer_id,
        st.opened_datetime
    FROM support_tickets st
    LEFT JOIN order_transactions ot ON st.order_id = ot.order_id
    WHERE st.restaurant_id IS NOT NULL
),

complaint_summary AS (
    SELECT 
        restaurant_id,
        COUNT(*) AS total_complaints,
        ROUND(AVG(resolution_time_min), 2) AS avg_resolution_time_min,
        SUM(COALESCE(compensation_amount, 0)) AS total_compensation
    FROM restaurant_complaints
    GROUP BY restaurant_id
),

complaint_types AS (
    SELECT 
        restaurant_id,
        issue_sub_type,
        COUNT(*) AS type_count
    FROM restaurant_complaints
    GROUP BY restaurant_id, issue_sub_type
),

restaurant_orders AS (
    SELECT 
        restaurant_id,
        COUNT(*) AS total_orders
    FROM order_transactions
    WHERE order_status = 'completed'
    GROUP BY restaurant_id
),

complaint_ratio AS (
    SELECT 
        cs.restaurant_id,
        cs.total_complaints,
        ro.total_orders,
        ROUND(CAST(cs.total_complaints AS FLOAT) / NULLIF(ro.total_orders, 0), 2) AS complaint_to_order_ratio,
        cs.avg_resolution_time_min,
        cs.total_compensation
    FROM complaint_summary cs
    LEFT JOIN restaurant_orders ro ON cs.restaurant_id = ro.restaurant_id
),

customer_latest_complaint AS (
    SELECT
        restaurant_id,
        customer_id,
        MAX(opened_datetime) AS latest_complaint_time
    FROM restaurant_complaints
    GROUP BY restaurant_id, customer_id
),

repeat_purchase_impact AS (
    SELECT 
        clc.restaurant_id,
        clc.customer_id,
        COUNT(*) AS repeat_orders_after_complaint
    FROM customer_latest_complaint clc
    JOIN order_transactions ot 
        ON clc.customer_id = ot.customer_id 
        AND clc.restaurant_id = ot.restaurant_id
        AND ot.order_datetime > clc.latest_complaint_time
    GROUP BY clc.restaurant_id, clc.customer_id
),

repeat_behavior_summary AS (
    SELECT 
        restaurant_id,
        ROUND(AVG(repeat_orders_after_complaint), 2) AS avg_repeat_orders_after_complaint
    FROM repeat_purchase_impact
    GROUP BY restaurant_id
)

SELECT
    cr.restaurant_id,
    cr.total_complaints,
    cr.total_orders,
    cr.complaint_to_order_ratio,
    cr.avg_resolution_time_min,
    cr.total_compensation,
    rbs.avg_repeat_orders_after_complaint,
    ct.issue_sub_type,
    ct.type_count
FROM complaint_ratio cr
LEFT JOIN repeat_behavior_summary rbs ON cr.restaurant_id = rbs.restaurant_id
LEFT JOIN complaint_types ct ON cr.restaurant_id = ct.restaurant_id
ORDER BY cr.complaint_to_order_ratio DESC NULLS LAST;


"""

df_restaurant_complaint_report = pd.read_sql_query(query_restaurant_complaint_report,con)
df_restaurant_complaint_report
```
